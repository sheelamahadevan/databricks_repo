```python
# find duplicate in df
print(cust_raw_df.count()) # action
#row level unique count
print(cust_raw_df.distinct().count()) # deduplication on record level transformation.action
print(cust_raw_df.dropDuplicates().count())

cust_raw_df.filter("custid='4000001'").show()
cust_raw_df.dropDuplicates().filter("custid='4000001'").show()
cust_raw_df.distinct().filter("custid='4000001'").show()
```
